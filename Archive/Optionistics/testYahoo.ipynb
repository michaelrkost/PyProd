```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Chrome linux User Agent - needed to not get blocked as a bot
headers = {
 'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}

symbol =  "FB"

print("getMarketData", symbol)

s = requests.Session()
aURL = "http://www.optionistics.com/quotes/stock-quotes/FB"
r = s.post(aURL, data={'s': symbol}, headers = headers)

r.close()
```

```python
r.content
```

```python

# get web page into BeautifulSoup
src = r.content
soup = BeautifulSoup(src, 'html.parser')
```

```python
print(soup)

```

```python

# Find the data in the table with the id: 'mainbody'
# and table class: 'quotem'
table = soup.find_all('table', {'class': 'quotem'})
#got it - now put it in a DF

df = pd.read_html(str(table))



```
